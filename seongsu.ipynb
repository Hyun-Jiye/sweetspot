```python
data1 = pd.read_csv('data/성수동/LOCAL_PEOPLE_DONG_202301.csv', encoding='utf-8-sig')
data2 = pd.read_csv('data/성수동/LOCAL_PEOPLE_DONG_202302.csv', encoding='utf-8-sig')
data3 = pd.read_csv('data/성수동/LOCAL_PEOPLE_DONG_202303.csv', encoding='utf-8-sig')
data4 = pd.read_csv('data/성수동/LOCAL_PEOPLE_DONG_202304.csv', encoding='utf-8-sig')
data5 = pd.read_csv('data/성수동/LOCAL_PEOPLE_DONG_202305.csv', encoding='utf-8-sig')
data6 = pd.read_csv('data/성수동/LOCAL_PEOPLE_DONG_202306.csv', encoding='utf-8-sig')
data7 = pd.read_csv('data/성수동/LOCAL_PEOPLE_DONG_202307.csv', encoding='utf-8-sig')
data8 = pd.read_csv('data/성수동/LOCAL_PEOPLE_DONG_202308.csv', encoding='utf-8-sig')
data9 = pd.read_csv('data/성수동/LOCAL_PEOPLE_DONG_202309.csv', encoding='utf-8-sig')
data10 = pd.read_csv('data/성수동/LOCAL_PEOPLE_DONG_202310.csv', encoding='utf-8-sig')
data11 = pd.read_csv('data/성수동/LOCAL_PEOPLE_DONG_202311.csv', encoding='utf-8-sig')
data12 = pd.read_csv('data/성수동/LOCAL_PEOPLE_DONG_202312.csv', encoding='utf-8-sig')
```

```python
d1 = pd.concat([data1, data2], axis=0)
d2 = pd.concat([d1, data3], axis=0)
d3 = pd.concat([d2, data4], axis=0)
d4 = pd.concat([d3, data5], axis=0)
d5 = pd.concat([d4, data6], axis=0)
d6 = pd.concat([d5, data7], axis=0)
d7 = pd.concat([d6, data8], axis=0)
d8 = pd.concat([d7, data9], axis=0)
d9 = pd.concat([d8, data10], axis=0)
d10 = pd.concat([d9, data11], axis=0)
data = pd.concat([d10, data12], axis=0)
```

### 주말, 평일, 주차 산정

```python
week = []     # 요일
ju = []       # 주차
    
for idx, row in a.iterrows():
    year = str(idx)[:4]
    month = str(idx)[4:6]
    day = str(idx)[6:8]
    weekday = date(int(year), int(month), int(day)).weekday()

    if weekday > 4:
        week.append('주말')
    else:
        week.append('평일')

    ju.append(datetime(int(year), int(month), int(day)).isocalendar()[1])
    
    print("행 인덱스: ", idx)
    print(row['행정동코드'], '\n')
```

```python
week
```

```python
ju
```

```python
seongsu['요일'] = week
seongsu['주차'] = ju
seongsu
```

### 요일별 통계

```python
mean_of_day = seongsu.groupby('요일')['행정동코드'].mean()
mean_of_day
```

```python
median_of_day = seongsu.groupby('요일')['행정동코드'].median()
median_of_day
```

```python
a = pd.DataFrame(mean_of_day)
b = pd.DataFrame(median_of_day)
```

```python
population_of_day = pd.concat([a,b], axis=1, join='inner')     # 데이터프레임 합치기
population_of_day
```

```python
population_of_day.columns = ['유동인구수(평균)', '유동인구수(중앙값)']   # 컬럼명 변경
population_of_day = population_of_day.astype('int')
population_of_day
```

### 주별 통계

```python
mean_of_week = seongsu.groupby('주차')['행정동코드'].mean()
mean_of_week
```

```python
median_of_week = seongsu.groupby('주차')['행정동코드'].median()
median_of_week
```

```python
population_of_week = pd.concat([c,d], axis=1, join='inner')     # 데이터프레임 합치기
population_of_week
```
