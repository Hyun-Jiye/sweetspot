```python
import pandas as pd
```

## 데이터 불러오기

```python
data1 = pd.read_csv('data/매장 관련/서울시 일반음식점 인허가 정보.csv', encoding='cp949')
data2 = pd.read_csv('data/매장 관련/서울시 휴게음식점 인허가 정보.csv', encoding='cp949')
data3 = pd.read_csv('data/매장 관련/서울시 제과점영업 인허가 정보.csv', encoding='cp949')
```

```python
data1 = data1.iloc[:,:26]
data2 = data2.iloc[:,:26]
data3 = data3.iloc[:,:26]
```

## 데이터프레임 병합

```python
d1 = pd.concat([data1, data2], axis=0)
data = pd.concat([d1, data3], axis=0)
```

```python
data
```

## 2019-01-01 이후 인허가 등록된 매장 추출

```python
data = data[data['인허가일자']>='2019-01-01'].sort_values('인허가일자')
data
```

## 지하 5층 - 지상 5층까지의 매장 추출

```python
data = data[(data['도로명주소'].str.contains('1층')==True) | (data['도로명주소'].str.contains('2층')==True) | (data['도로명주소'].str.contains('3층')==True) | (data['도로명주소'].str.contains('4층')==True) | (data['도로명주소'].str.contains('5층')==True)]
data
```

## 층별 비중 확인

```python
sectors = data.value_counts('업태구분명').to_frame()
sectors = sectors.reset_index()
sectors
```

```python
list_1 = list(sectors['업태구분명'])
list_2 = list(sectors['count'])
d = dict(zip(list_1, list_2))
d
```

```python
floor_1 = data[(data['도로명주소'].str.contains('1층')==True) & -(data['도로명주소'].str.contains('지하1층'))]
floor_1_sectors = floor_1.groupby('업태구분명').size().to_frame().reset_index()
floor_1_sectors.columns = ['업종', '1층 점포 수']
floor_1_sectors['조사 표본 수'] = floor_1_sectors['업종'].map(d)
floor_1_sectors['1층 비중(%)'] = round(floor_1_sectors['1층 점포 수']/floor_1_sectors['조사 표본 수']*100, 2)
floor_1_sectors = floor_1_sectors.sort_values('1층 비중(%)', ascending=False)
floor_1_sectors
```
