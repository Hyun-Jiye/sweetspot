```python
import requests
import pprint 
import json 
import pandas as pd
```

```python
df = pd.DataFrame()
page_no = 1

while True:
    url = "http://apis.data.go.kr/1130000/FftcBrandFrcsDropInfoService/getBrandFrcsDropInfo?serviceKey=Xp8aQvdAcrDsMdFYBMaaa%2F%2B%2Bplq6bhwzvgxJynJrgt8qIh7wxfHuOUVLThJ3ilOVJ1tuEydVFIRZkQlaN3ReAQ%3D%3D&pageNo={0}&numOfRows=10000&resultType=json&yr=2022".format(page_no)
    response = requests.get(url)
    contents = response.text
    json_db = json.loads(contents)
    doc = json_db["items"]
    df = pd.concat([df, pd.DataFrame(doc)] ,axis = 0 , ignore_index = True)

    print(url)
    print(page_no)
    page_no += 1
    if page_no == 18 : 
        break 
        
    
```
