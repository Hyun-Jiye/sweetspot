```python
df = pd.read_csv('data/법정동코드.csv', encoding='UTF8')
```

```python
df['법정동코드'] = df['법정동코드'].astype('string')
df = df[(df['법정동코드'].str.endswith('000000')) & (df['폐지여부']=='존재')]
locations = df['법정동코드'].tolist()
locations = locations[1:]
locations
```

```python
months = ['202301', '202302', '202303', '202304', '202305', '202306', '202307', '202308', '202309', '202310', '202311', '202312']
```

```python
ages = ['10', '20', '30']
```

```python
headers = {
    "accept": "application/json",
    "Content-Type": "application/json",
    "appkey": "naEzoo9JOf3MBzRYgVYhkiHwc207eHb3aZcDekjf"
}

response = requests.get(url, headers=headers)

print(response.text)
```

```python
# 초기 데이터프레임 설정
columns = ['districtCode', 'districtName', 'gender', 'ageGrp', 'companionType', 'travelerCount', 'yearMonth']
all_data = pd.DataFrame(columns=columns)

# 반복문을 사용하여 API 호출
for location in locations:
    for month in months:
        for age in ages:
            url = "https://apis.openapi.sk.com/puzzle/travel/visit/count/raw/monthly/districts/{0}?yearMonth={1}&gender=all&ageGrp={2}".format(location, month, age)
            headers = {
                "accept": "application/json",
                "Content-Type": "application/json",
                "appkey": "naEzoo9JOf3MBzRYgVYhkiHwc207eHb3aZcDekjf"
            }

            # API 요청 및 응답
            response = requests.get(url, headers=headers)
            if response.status_code == 200:
                data = response.json()
                if data.get('status', {}).get('code') == '00':  # 정상 응답인 경우
                    content = data['contents']
                    row = {
                        'districtCode': content['districtCode'],
                        'districtName': content['districtName'],
                        'gender': content['gender'],
                        'ageGrp': content['ageGrp'],
                        'companionType': content['companionType'],
                        'travelerCount': content['raw']['travelerCount'],
                        'yearMonth': content['raw']['yearMonth']
                    }
                    # 결과 데이터프레임에 추가
                    all_data = pd.concat([all_data, pd.DataFrame([row])], ignore_index=True)
            else:
                print(f"Failed to fetch data for {location}, {month}, {age}. Status code: {response.status_code}")
                
print(all_data)
```
