```python
import pandas as pd
```

```python
data = pd.read_excel('data/광고성과_0701.xlsx')
```

```python
data
```

```python
# 날짜 형식을 datetime으로 변환
data['날짜'] = pd.to_datetime(data['날짜'], format='%Y.%m.%d')

# 광고 시작 날짜로부터 며칠째인지를 계산
data['광고 시작 후 일수'] = data.groupby(['매체', '광고 세트', '소재'])['날짜'].rank(method='first')

# 1일차 데이터를 건너뛰고 분석
filtered_data = data[data['광고 시작 후 일수'] > 1]

# 결과당 비용의 변화를 계산
filtered_data['결과당 비용 변화'] = filtered_data.groupby(['매체', '광고 세트', '소재'])['결과당 비용'].diff()

# 결과당 비용이 증가하는 시점을 찾기 위해 변화가 양수인 시점을 선택
increasing_cost_points_filtered = filtered_data[filtered_data['결과당 비용 변화'] > 0]

# 광고 시작 후 결과당 비용이 증가하는 첫 번째 시점을 계산
first_increase_points = increasing_cost_points_filtered.groupby(['매체', '광고 세트', '소재']).agg({'광고 시작 후 일수': 'min'}).reset_index()

# 평균 일수 계산
average_increasing_cost_day_filtered = first_increase_points['광고 시작 후 일수'].mean()

# 결과 데이터프레임 출력
first_increase_points, average_increasing_cost_day_filtered
```
