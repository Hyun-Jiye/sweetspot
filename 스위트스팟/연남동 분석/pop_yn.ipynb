```python
import pandas as pd
import glob
import chardet
import os

# 파일들이 있는 폴더의 경로
folder_paths = ['./연남유동인구데이터/2019', './연남유동인구데이터/2020', './연남유동인구데이터/2021', './연남유동인구데이터/2022', './연남유동인구데이터/2023']
combined_df = pd.DataFrame()

# 필터링할 행정동 코드
filter_codes = ['11440710', '11440660']

for folder_path in folder_paths:
    all_files = glob.glob(os.path.join(folder_path, "*.csv"))
    for filename in all_files:
        # 파일의 인코딩 감지
        with open(filename, 'rb') as f:
            result = chardet.detect(f.read(100000))
            encoding = result['encoding']

        # 파일 읽기, 첫 번째 행(헤더)을 무시
        df = pd.read_csv(filename, encoding=encoding, dtype='object')

        # 2019년 10월 파일을 제외하고 처리
        if '201910' not in filename:
            # 데이터프레임을 풀어주고 컬럼명 덮어쓰기
            df.reset_index(inplace=True)
            # 덮어씌울 새로운 컬럼명 정의
            new_columns = ['기준일ID', '시간대구분', '행정동코드', '총생활인구수'] + [f'extra_{i}' for i in range(df.shape[1]-4)]
            df.columns = new_columns
        else:
            # 2019년 10월 데이터는 처음 4개의 컬럼만 사용
            df = df.iloc[:, :4]
            df.columns = ['기준일ID', '시간대구분', '행정동코드', '총생활인구수']

        # 행정동코드로 필터링
        df['행정동코드'] = df['행정동코드'].astype(str)
        df = df[df['행정동코드'].isin(filter_codes)]

        # 병합 및 정렬 
        combined_df = pd.concat([combined_df, df], ignore_index=True)
        combined_df = combined_df.sort_values(by=['기준일ID'],ascending=True)

# 결과 확인
print(combined_df.head())
combined_df.to_csv('final_combined_files.csv', index=False, encoding='utf-8-sig')

combined_df.head()
```

```python
# 시간대별로는 다 합하고, (유동인구의 경우 밤시간대 혹은 새벽시간대에 극단값을 보일 수 있으므로)
# 일자별로 묶어줌.
# quarters로 년도별 - 분기별 컬럼으로 가공시켜준다. 

combined_df = combined_df.iloc[:,:4]


combined_df['총생활인구수'] = combined_df['총생활인구수'].astype('float')

combined_df.dtypes

# combined_df.groupby('기준일ID')['총생활인구수'].sum().reset_index()
```

```python
combined_df['기준일ID'] = pd.to_datetime(combined_df['기준일ID'])

combined_df['year'] = combined_df['기준일ID'].dt.year
combined_df['Q'] = combined_df['기준일ID'].dt.quarter


# df_grouped = combined_df.groupby(['year','Q'])['총생활인구수'].mean().reset_index()
# df_grouped['year_Q'] = df_grouped['year'].astype(str) + 'Q' + df_grouped('Q').astype(str)


```

```python
df_grouped = combined_df.groupby(['year','Q'])['총생활인구수'].mean().reset_index()
```

```python
df_grouped['year_Q'] = df_grouped['year'].astype(str) + 'Q' + df_grouped['Q'].astype(str)
```

```python
df_grouped = df_grouped.iloc[:,2:]

new_orders = ['year_Q', '총생활인구수']


df_grouped = df_grouped[new_orders]
```

```python
df_grouped.to_csv('분기별 총생활인구수.csv', encoding='utf-8-sig', index=False)


```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.rc('font', family = 'AppleGothic') #font 설정
plt.rc('axes', unicode_minus = False) #마이너스 기호 깨짐 문제 방지
 
plt.figure(figsize=(12,8))
sns.lineplot(x= 'year_Q', y='총생활인구수', data = df_grouped,marker = 'o')
plt.xticks(rotation = 45)
plt.title('분기별 총생활인구수 시계열 변화')
plt.xlabel('년도-분기')
plt.ylabel('총생활인구수')

df_grouped['difference'] = df_grouped['총생활인구수'].diff() #먼저 변화량을 계산 
threshold = df_grouped['difference'].quantile(0.05) #임계값 (변화량의 10번째 백분위수를 임계값으로 설정한다.) 
significant_drops = df_grouped[df_grouped['difference'] <= threshold]

for index , row in significant_drops.iterrows():
    plt.annotate('급감 지점', (row['year_Q'], row['총생활인구수']),
                 textcoords='offset points', xytext = (-15,-10),ha = 'center', color = 'red')



plt.tight_layout()
plt.show()
```
