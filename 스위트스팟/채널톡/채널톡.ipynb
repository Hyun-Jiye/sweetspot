```python
import requests
import pandas as pd
import json
from datetime import datetime
import pytz
```

## 1차 테스트

### 목적 : 고객 이벤트 데이터 API로 수집 (특정 고객 1명)

```python
# 1. 채널톡 API 엔드포인트 및 인증 정보
api_url = 'https://api.channel.io/open/v5/users/66455f37334f5868d186/events'
access_key = '665ea42596ce809bd8e8'
access_secret = '6b984ddd146893933457a35dfb083842'
```

```python
# 헤더 설정
headers = {
    'accept': 'application/json',
    'x-access-key': access_key,
    'x-access-secret': access_secret
}
```

```python
# 2. API 요청 보내기
response = requests.get(api_url, headers=headers)

if response.status_code == 200:
    # 응답 텍스트를 UTF-8로 디코딩
    response.encoding = 'utf-8'
    data = response.json()
    
    # 데이터 구조 확인
    print(json.dumps(data, indent=2, ensure_ascii=False))
    
    # 3. 데이터를 데이터프레임으로 변환
    if 'events' in data and isinstance(data['events'], list):
        # 이벤트 리스트를 데이터프레임으로 변환
        events_df = pd.json_normalize(data['events'])
        
        # 타임스탬프 필드를 사람이 읽을 수 있는 형식으로 변환
        # UTC로 변환
        events_df['createdAt'] = pd.to_datetime(events_df['createdAt'], unit='ms', utc=True)
        events_df['expireAt'] = pd.to_datetime(events_df['expireAt'], unit='ms', utc=True)
        events_df['property.session'] = pd.to_datetime(events_df['property.session'], unit='ms', utc=True)
        
        # 타임존 변환 (한국 표준시로 변환)
        kst = pytz.timezone('Asia/Seoul')
        events_df['createdAt'] = events_df['createdAt'].dt.tz_convert(kst)
        events_df['expireAt'] = events_df['expireAt'].dt.tz_convert(kst)
        events_df['property.session'] = events_df['property.session'].dt.tz_convert(kst)
        
        # 4. CSV 파일로 저장
        events_df.to_csv('channeltalk_events.csv', index=False, encoding='utf-8-sig')
    
    else:
        print("The 'events' key is not present or is not a list.")

else:
    print(f"Failed to retrieve data: {response.status_code}")
```

## 2차 테스트

### 목적 : 고객 이벤트 데이터 API로 수집 (전체 고객)

```python
# 전체 고객 데이터 로드
customer = pd.read_excel('data/채널톡 고객 데이터(2024-06-04 16시).xlsx')
```

```python
# 병합을 위한 컬럼명 변경 
customer.rename(columns={'id':'userId'}, inplace=True)
customer
```

```python
# 고객 ID 리스트화
user_list = list(customer['userId'])
user_list
```

```python
# 빈 데이터프레임 생성
all_events_df = pd.DataFrame()

# 한국 표준시 타임존 설정
kst = pytz.timezone('Asia/Seoul')

# 반복문을 통해 각 사용자에 대해 API 호출
for user in user_list:
    # 1. 채널톡 API 엔드포인트 및 인증 정보
    api_url = f'https://api.channel.io/open/v5/users/{user}/events'
    access_key = '665ea42596ce809bd8e8'
    access_secret = '6b984ddd146893933457a35dfb083842'
    
    # 헤더 설정
    headers = {
        'accept': 'application/json',
        'x-access-key': access_key,
        'x-access-secret': access_secret
    }
    
    # API 요청 보내기
    response = requests.get(api_url, headers=headers)
    
    print(f"Response status code for user {user}: {response.status_code}")
    
    if response.status_code == 200:
        # 응답 텍스트를 UTF-8로 디코딩
        response.encoding = 'utf-8'
        data = response.json()
        
        # 데이터 구조 확인
        print(json.dumps(data, indent=2, ensure_ascii=False))
        
        # 데이터를 데이터프레임으로 변환
        if 'events' in data and isinstance(data['events'], list) and len(data['events']) > 0:
            events_list = []
            for event in data['events']:
                # 'createdAt' 및 'expireAt' 필드 확인 및 기본값 설정
                event['createdAt'] = event.get('createdAt', None)
                event['expireAt'] = event.get('expireAt', None)
                if 'property' in event and isinstance(event['property'], dict):
                    event['property.session'] = event['property'].get('session', None)
                else:
                    event['property.session'] = None
                events_list.append(event)
            
            events_df = pd.json_normalize(events_list)
            
            # 타임스탬프 필드를 사람이 읽을 수 있는 형식으로 변환
            if 'createdAt' in events_df.columns:
                events_df['createdAt'] = pd.to_datetime(events_df['createdAt'], unit='ms', utc=True).dt.tz_convert(kst)
            if 'expireAt' in events_df.columns:
                events_df['expireAt'] = pd.to_datetime(events_df['expireAt'], unit='ms', utc=True).dt.tz_convert(kst)
            if 'property.session' in events_df.columns:
                events_df['property.session'] = pd.to_datetime(events_df['property.session'], unit='ms', utc=True).dt.tz_convert(kst)
            
            # 결과를 누적
            all_events_df = pd.concat([all_events_df, events_df], ignore_index=True)
        else:
            print(f"No events found for user: {user}, {user_list.index(user)}번째")
            continue
    else:
        print(f"Failed to retrieve data for user {user}: {response.status_code}")
        print(f"Response content for user {user}: {response.content}")

# 최종 데이터프레임을 CSV 파일로 저장
if not all_events_df.empty:
    all_events_df.to_csv('channeltalk_all_events.csv', index=False, encoding='utf-8-sig')
else:
    print("No events data was retrieved.")
```

```python
# 고객 데이터 필요한 컬럼만 추출
customer_df = customer[['userId', 'name', 'profile.office']]
customer_df
```

```python
# 이벤트 데이터와 고객 데이터 병합
merged_df = pd.merge(all_events_df, customer_df, on='userId', how='left')
merged_df
```

```python
# 멀티인덱스 설정
merged_df = merged_df.set_index(['userId', 'id'])
merged_df
```
