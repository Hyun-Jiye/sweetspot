```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm
from matplotlib.ticker import ScalarFormatter
```

```python
data1 = pd.read_csv('data/상권 관련/서울시_상권분석서비스(추정매출-상권)_2019년.csv', encoding='cp949', index_col=None)
data2 = pd.read_csv('data/상권 관련/서울시_상권분석서비스(추정매출-상권)_2020년.csv', encoding='cp949', index_col=None)
data3 = pd.read_csv('data/상권 관련/서울시_상권분석서비스(추정매출-상권)_2021년.csv', encoding='cp949', index_col=None)
data4 = pd.read_csv('data/상권 관련/서울시_상권분석서비스(추정매출-상권)_2022년.csv', encoding='cp949', index_col=None)
data5 = pd.read_csv('data/상권 관련/서울시_상권분석서비스(추정매출-상권)_2023년.csv', encoding='cp949', index_col=None)
```

```python
data1 = data1[data1['상권_코드_명']=='가로수길']
data2 = data2[data2['상권_코드_명']=='가로수길']
data3 = data3[data3['상권_코드_명']=='가로수길']
data4 = data4[data4['상권_코드_명']=='가로수길']
data5 = data5[data5['상권_코드_명']=='가로수길']
```

```python
d1 = pd.concat([data1, data2], axis=0)
d2 = pd.concat([d1, data3], axis=0)
d3 = pd.concat([d2, data4], axis=0)
data = pd.concat([d3, data5], axis=0)
```

```python
count_per_quarter = data.groupby('기준_년분기_코드')['서비스_업종_코드'].size()

print(count_per_quarter)
```

```python
average_sales_per_quarter = data.groupby('기준_년분기_코드')['당월_매출_금액'].mean()
print(average_sales_per_quarter)
```

```python
average_sales_per_quarter = pd.DataFrame(average_sales_per_quarter).reset_index()
average_sales_per_quarter
```

```python
# 한글 폰트 설정
plt.rc('font', family='Malgun Gothic') 

# '기준_년분기_코드'를 문자열로 변환 후 '연도-분기' 형식으로 변환
data['기준_년분기_코드'] = data['기준_년분기_코드'].apply(lambda x: f"{str(x)[:-1]}Q{str(x)[-1]}")
```

```python
# 기준 년분기별 매출 금액 합계 계산
sales_sum_per_quarter = data.groupby('기준_년분기_코드')['당월_매출_금액'].sum()

# 시계열 꺾은선 그래프로 시각화
plt.figure(figsize=(10, 6))  # 그래프 크기 설정
plt.plot(sales_sum_per_quarter.index, sales_sum_per_quarter.values, marker='o', linestyle='-', color='blue')  # 꺾은선 그래프 생성
plt.xlabel('기준 년분기')  # x축 레이블
plt.ylabel('매출 금액 합계')  # y축 레이블
plt.title('기준 년분기별 매출 금액 합계 추이')  # 그래프 제목

# Y축을 정수 형식으로 설정
y_formatter = ScalarFormatter(useOffset=False)
y_formatter.set_scientific(False)
plt.gca().yaxis.set_major_formatter(y_formatter)

plt.grid(True)  # 격자 표시
plt.xticks(rotation=45)  # x축 레이블 회전
plt.tight_layout()  # 레이아웃 조정
plt.show()  # 그래프 표시
```

```python
# 기준 년분기별 매출 금액 평균 계산
sales_avg_per_quarter = data.groupby('기준_년분기_코드')['당월_매출_금액'].mean()

# 막대그래프로 시각화
plt.figure(figsize=(10, 6))  # 그래프 크기 설정
plt.bar(sales_avg_per_quarter.index, sales_avg_per_quarter.values, color='skyblue')  # 막대그래프 생성
plt.xlabel('기준 년분기')  # x축 레이블
plt.ylabel('평균매출 금액')  # y축 레이블
plt.title('기준 년분기별 평균 월매출 금액 추이')  # 그래프 제목

# Y축을 정수 형식으로 설정
y_formatter = ScalarFormatter(useOffset=False)
y_formatter.set_scientific(False)
plt.gca().yaxis.set_major_formatter(y_formatter)

plt.grid(True)  # 격자 표시
plt.xticks(rotation=45)  # x축 레이블 회전
plt.tight_layout()  # 레이아웃 조정
plt.show()  # 그래프 표시
```

```python
data1 = pd.read_csv('data/상권 관련/서울시_상권분석서비스(점포-상권)_2019년.csv', encoding='cp949', index_col=None)
data2 = pd.read_csv('data/상권 관련/서울시_상권분석서비스(점포-상권)_2020년.csv', encoding='cp949', index_col=None)
data3 = pd.read_csv('data/상권 관련/서울시_상권분석서비스(점포-상권)_2021년.csv', encoding='cp949', index_col=None)
data4 = pd.read_csv('data/상권 관련/서울시_상권분석서비스(점포-상권)_2022년.csv', encoding='cp949', index_col=None)
data5 = pd.read_csv('data/상권 관련/서울시_상권분석서비스(점포-상권)_2023년.csv', encoding='cp949', index_col=None)
```

```python
data1 = data1[data1['상권_코드_명']=='가로수길']
data2 = data2[data2['상권_코드_명']=='가로수길']
data3 = data3[data3['상권_코드_명']=='가로수길']
data4 = data4[data4['상권_코드_명']=='가로수길']
data5 = data5[data5['상권_코드_명']=='가로수길']
```

```python
d1 = pd.concat([data1, data2], axis=0)
d2 = pd.concat([d1, data3], axis=0)
d3 = pd.concat([d2, data4], axis=0)
df = pd.concat([d3, data5], axis=0)
```

```python
df
```

```python
# '기준_년분기_코드'와 '서비스_업종_코드_명' 별로 그룹화하고, 각 컬럼의 합계 계산
summary_df = df.groupby(['기준_년분기_코드', '서비스_업종_코드_명']).agg({
    '점포_수': 'sum',
    '개업_점포_수': 'sum',
    '폐업_점포_수': 'sum',
    '프랜차이즈_점포_수': 'sum'
})

summary_df = pd.DataFrame(summary_df).reset_index()
summary_df
```
