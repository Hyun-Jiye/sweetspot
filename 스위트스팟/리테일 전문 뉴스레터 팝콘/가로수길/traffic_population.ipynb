```python
import matplotlib.pyplot as plt 
import matplotlib_inline as mpl


plt.rcParams['font.family'] = 'AppleGothic' #맥 os한글 폰트(기본)
plt.rcParams['axes.unicode_minus'] = False #마이너스 부호 표시 문제를 해결 


plt.figure(figsize=(12,8))
data_karosu_quartly['총_유동인구_수'].plot(kind='bar')
plt.title('분기별 총 유동인구수')
plt.xlabel('분기')
plt.ylabel('총 유동 인구 수')
plt.xticks(rotation = 45)
plt.show()
```

```python
data_karosu_quartly.sort_index(inplace=True)

data_karosu_quartly['MZ_유동인구_수'] = data_karosu_quartly['연령대_20_유동인구_수'] + data_karosu_quartly['연령대_30_유동인구_수']
data_karosu_quartly['MZ_유동인구_수'].plot(kind='line', figsize=(10,5), title = 'MZ세대 유동인구 수 추이', marker = 'o')


plt.xlabel('기준 년도 및 분기')
plt.ylabel('유동인구 수')
plt.grid(True)
plt.show()


# mz세대 유동인구수가 줄고 있다는 가설에 대해 통계적 유의성 검정 
# 전체 유동인구 수와 비교 

# 단순회귀분석을 통해 ,시간을 독립변수로 하고 유동인구를 종속변수로 하여 선형회귀모델을 사용해보자. 이를 통해 시간에 따른 경향성을 수치적으로 평가
# Dickey-Fuller검정을 통해 시계열이 정상성을 가지는지 검정하고 정상성이 없는 경우 즉 시계열에 단위근이 있는 경우 시간에 따라 유동인구 수가 추세적으로 감소 혹은 증가할 가능성이 높음.
# Mann-Kendall검정 : 비모수적 검정으로 데이터의 추세를 확인하는데 사용되며 데이터의 분포가 정규분포를 따르지 않을 때 유용하다. 

# 우리는 단순선형회귀모델을 사용하여 시간에 따른 경향성을 평가 
```

```python
import statsmodels.api as sm
import numpy as np
from sklearn.linear_model import LinearRegression
data_karosu_quartly['Time'] = np.arange(len(data_karosu_quartly.index))

X = data_karosu_quartly[['Time']]
y = data_karosu_quartly['MZ_유동인구_수']

model = LinearRegression().fit(X,y)

print('Slope : ', model.coef_[0]) # 기울기
print('Intercept: ', model.intercept_) # 절편

X2 = sm.add_constant(X)
est = sm.OLS(y ,X2)
est2 = est.fit()
print(est2.summary())


plt.scatter(X,y,color = 'blue')#데이터의 점
plt.plot(X, model.predict(X), color = 'red')#회귀선
plt.title('MZ세대 유동인구 수 시간에 따른 추세')
plt.xlabel('시간(인덱스 순서)')
plt.ylabel('MZ세대 유동인구 수 ')
plt.show()
```

#### 

선형모델의 기울기가 약 -656.30으로 시간이 지날수록 MZ세대의 유동인구 수가 감소하고 있다는 추세를 나타낼 수는 있다. 그러나 p-value가 0.804로 일반적인 유의수준 (0.05 또는 0.01)에서 이 추세가 통계적으로 유의하지 않다는 것을 의미하기 때문에 이러한 유동인구 감소 추세가 가설을 정당화할 수는 없음. 


```python
quarterly_data = data_karosu.groupby('기준_년분기_코드')['총_유동인구_수'].sum().reset_index()

quarterly_data['변화율'] = quarterly_data['총_유동인구_수'].pct_change() * 100 

threshold = quarterly_data['변화율'].std() * 2
upper_threshold = quarterly_data['변화율'].mean() + threshold
lower_threshold = quarterly_data['변화율'].mean() - threshold


change_part = (quarterly_data['변화율'] > upper_threshold) | (quarterly_data['변화율'] < lower_threshold)

plt.figure(figsize=(10,8))
plt.bar(quarterly_data['기준_년분기_코드'],quarterly_data['총_유동인구_수'], color = 'gray')
plt.title('분기별 총 유동 인구수')
plt.xlabel('분기')
plt.ylabel('총 유동 인구 수')

plt.bar(quarterly_data[change_part]['기준_년분기_코드'], quarterly_data[change_part]['총_유동인구_수'], color = 'red')
plt.xticks(rotation = 45)

for x,y,pct in zip(quarterly_data['기준_년분기_코드'],quarterly_data['총_유동인구_수'],quarterly_data['변화율']):
    if pct > upper_threshold or pct < lower_threshold :
        plt.text(x,y,f'{pct:.2f}%', ha = 'center', va = 'bottom', color = 'black')
plt.show()
```

```python
## date 형식으로 년-분기값을 Transform시켜줌 
def convert_quarter(date): 
    year = int(str(date)[:4])
    quarter = int(str(date)[-1])

    month = 3 * (quarter - 1) + 1

    return pd.Timestamp(year = year, month =month , day=1)


total_df_1['기준_년분기_코드'] = total_df_1['기준_년분기_코드'].apply(convert_quarter)
total_df_1.set_index('기준_년분기_코드', inplace= True)


# 상권 지표 인코딩 
mapping = {'LL' : 0 ,'LH' : 1,'HL' : 2,'HH': 3}
total_df_1['상권지표_인코딩값'] = total_df_1['상권_변화_지표'].map(mapping)


plt.figure(figsize = (10,8))
plt.plot(total_df_1.index , total_df_1['상권지표_인코딩값'] , label = '인코딩된 상권지표')
plt.title('인코딩된 상권지표 시계열')
plt.xlabel('날짜')
plt.ylabel('상권지표')
plt.legend()
plt.show()


```

```python
X = total_df_1[['총_유동인구_수']]
y = total_df_1['상권지표_인코딩값']

X = sm.add_constant(X)
model = sm.OLS(y,X).fit()

print(model.summary())

#상권지표가 동일한 값이기 때문에 정상적인 통계량 측정이 되지 않음 
# R-squared(결정계수) : 모델이 데이터를 얼마나 잘 설명하는지를 나타낸다. 값이 높을수록 모델의 설명력이 높다고 볼 수 있다.
# coef(계수) : 유동인구 수가 1 증가할 때 상권 변화 지표의 변화량을 나타내는 계수 (계수가 양수이면 유동인구 수가 많을수록 해당 상권이 확장 상승하는 경향이 있다는 것을 의미한다.)

# P >|t|(p-value) : 계수의 유의성을 검정. 일반적으로 0.05 미만일 경우, 통계적으로 유의하다고 판별. 
```

### 해석결과

- R-squared 값은 0.220 으로 나타나 임대료가 공실률의 약 22퍼센트를 설명할 수 있음을 보여준다. 
- 회귀계수의 경우, 임대료에 대해 약 -1.3433이며 이는 임대료가 1만큼 증가할 때 공실률이 평균적으로 1.3433% 만큼 감소한다는 것을 의미 
- 이 계수는 통계적으로 유의미하며 (p-value : 0.043) 임대료의 증가가 공실률 감소에 유의미한 영향을 미친다고 해석할 수 있다.
- F통계량 같은 경우, 4.803이며 이는 모델이 통계적으로 유의하다는 것을 나타낸다. 그러나 R-squared값이 1에 가까울수록 설명력이 높다는 점을 고려할때 다른 요인 또한 공실률에 영향을 미칠 수 있으므로 추가적인 변수를 고려한 모델 구축의 필요성이 있음. (더 조사해 볼 필요성이 있다.)
분석결과에서는 결과적으로 임대료와 공실률 간의 음의 선형 관계가 존재함을 확인하였으며 이는 임대료가 상승하면 공실률이 감소하는 경향이 있음을 시사한다.

회귀계수 -0.4694는 공실률이 1표준편차가 증가할때 임대료가 0.4694 표준편차 감소함을 의미하며 이는 통계적으로 유의미하다. p-value : 0.043

결과 해석 : 공실률이 증가하면 임대료가 감소할 가능성이 있음을 시사하고 ,이러한 관계는 부동산 시장에서 공급과 수요의 법칙을 반영할 수 있음. 즉, 공실률이 높아지면 상대적으로  임대공간에 대한 수요가 낮아져 임대료를 낮춰야 할 필요성이 생긴다.
