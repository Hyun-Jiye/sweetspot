```python
from datetime import datetime, date
```

```python
week = []
ju = []

for idx, row in total_of_day.iterrows():
    year = str(idx)[:4]
    month = str(idx)[4:6]
    day = str(idx)[6:]
    weekday = date(int(year), int(month), int(day)).weekday()
    
    if weekday > 4 :
        week.append('주말')
        
    else:
        week.append('평일')
    ju.append(datetime(int(year), int(month), int(day)).isocalendar()[1])
    print('행 인덱스 : ',idx)
    print(row['총생활인구수'], '\n')
    
```

```python
print(week, ju)

```

```python
total_of_day['요일'] = week
total_of_day['주차'] = ju 

```

```python
### 통계량 확인 

mean_of_day = total_of_day.groupby('요일')['총생활인구수'].mean()
median_of_day = total_of_day.groupby('요일')['총생활인구수'].median()
mean_of_week = total_of_day.groupby('주차')['총생활인구수'].mean()
median_of_week = total_of_day.groupby('주차')['총생활인구수'].median()
```
