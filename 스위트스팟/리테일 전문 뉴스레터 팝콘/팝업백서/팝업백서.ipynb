## 지역 관련 처리

### 지역 리스트 생성

```python
seongsu = ['성수1가1동', '성수1가2동', '성수2가1동', '성수2가3동']
hongdae = ['서교동', '연남동']
jongro = ['종로1·2·3·4가동', '종로5·6가동']
hannam = ['한남동']
jamsil = ['잠실본동', '잠실2동', '잠실3동', '잠실6동', '삼전동', '석촌동', '송파1동']
gangnam = ['서초1동', '서초2동', '서초4동', '도곡1동', '역삼1동', '논현1동', '반포1동']
yeouido = ['여의동', '영등포동', '당산1동', '당산2동']
yongsan = ['한강로동']
```

```python
# 지역 맵핑 딕셔너리 생성
region_mapping = {
    '성수': seongsu,
    '홍대': hongdae,
    '종로': jongro,
    '한남': hannam,
    '잠실': jamsil,
    '강남': gangnam,
    '여의도': yeouido,
    '용산': yongsan
}
```

### 지역 매핑 함수 생성

```python
# 행정동_코드_명에 따른 지역을 매핑하는 함수 생성
def get_region(행정동_코드_명):
    for region, codes in region_mapping.items():
        if 행정동_코드_명 in codes:
            return region
    return None
```

### 지역별 면적 데이터프레임 생성

```python
# 지역별 면적 합계를 계산하는 함수
def get_region_area(df, region_name, area_list):
    region_df = df[df['행정동_명'].isin(area_list)]
    total_area = region_df['영역_면적'].sum()
    return {'지역': region_name, '면적': total_area}

# 새로운 데이터프레임 생성
regions = []
for region_name, area_list in region_mapping.items():
    region_info = get_region_area(data4, region_name, area_list)
    regions.append(region_info)

area_df = pd.DataFrame(regions)
area_df['면적(㎢)'] = area_df['면적'] / 1_000_000

# 결과 출력
area_df
```

### 비율 시각화

```python
# '기준_년분기_코드'가 '20234'인 데이터 필터링
filtered_data = filtered_data_1[filtered_data_1['기준_년분기_코드'] == 20234]

# 지역별로 남성 및 여성 유동인구 수 비율 시각화
regions = filtered_data['지역'].unique()

for region in regions:
    region_data = filtered_data[filtered_data['지역'] == region]
    
    # 남성 유동인구 수와 여성 유동인구 수의 비율 계산
    total_male = region_data['남성_유동인구_수'].sum()
    total_female = region_data['여성_유동인구_수'].sum()
    
    # 비율 시각화
    labels = ['남성', '여성']
    sizes = [total_male, total_female]
    
    fig, ax = plt.subplots(figsize=(8, 6))
    ax.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90)
    ax.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
    ax.set_title(f'{region} 성별 유동인구 비율 (2023년 4분기)')
    
    plt.show()
```

### 지역별 유동인구 밀집도 

```python
# '기준_년분기_코드'와 '지역' 컬럼을 기준으로 '총_유동인구_수' 합계 구하기
result_1 = filtered_data_1.groupby(['기준_년분기_코드', '지역'])['총_유동인구_수'].sum().reset_index()

# 결과 출력
result_1
```

```python
# 면적 데이터와 '지역' 컬럼 기준으로 매핑
merged_df_1 = pd.merge(result_1, area_df, on='지역')

# '인구_밀집도' 컬럼 추가
merged_df_1['유동인구_밀집도'] = merged_df_1['총_유동인구_수'] / merged_df_1['면적(㎢)']

# 결과 출력
merged_df_1
```

```python
merged_df_1['기준_년분기_코드'] = merged_df_1['기준_년분기_코드'].astype('str')
```

```python
# 그래프 시각화
plt.figure(figsize=(12, 8))

for region in merged_df_1['지역'].unique():
    region_data = merged_df_1[merged_df_1['지역'] == region]
    plt.plot(region_data['기준_년분기_코드'], region_data['유동인구_밀집도'], marker='o', label=region)

plt.xlabel('기준 년분기 코드')
plt.ylabel('유동인구 밀집도 (유동인구 수 / 지역 면적)')
plt.title('년분기+지역별 유동인구 밀집도')
plt.legend(title='지역')
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()

# 그래프 출력
plt.show()
```
