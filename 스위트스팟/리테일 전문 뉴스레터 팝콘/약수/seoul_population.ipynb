```python
import pandas as pd 

df = pd.read_csv('./행정동별_MZ세대_추이.csv', encoding='utf-8')
```

```python
df.head()
```

```python
import matplotlib as mpl
import matplotlib.pyplot as plt 
import numpy as np 
mpl.rcParams['font.family'] = 'AppleGothic'
mpl.rcParams['axes.unicode_minus'] = False
```

```python
dong_gr = df['행정동_명'].unique()
dong_gr
```

```python
df['기준_년분기_코드'] = df['기준_년분기_코드'].apply(lambda x : f'{str(x)[:4]} Q{str(x)[4:]}')

df['기준_년분기_코드']
```

```python
plt.figure(figsize = (10,8))

for dong in dong_gr :
    dong_data = df[df['행정동_명'] == dong]
    plt.plot(dong_data['기준_년분기_코드'], dong_data['MZ세대'], label = dong)

plt.xlabel('기준 년분기')
plt.ylabel('MZ세대 유동인구수')
plt.title('분기별 행정동별 MZ세대 유동인구 추이')
plt.xticks(rotation = 45)
plt.legend(loc='upper left', bbox_to_anchor = (1,1), title = '행정동')
plt.grid(True, linestyle = '--', alpha= 0.7) 
plt.tight_layout()


plt.show()
```
