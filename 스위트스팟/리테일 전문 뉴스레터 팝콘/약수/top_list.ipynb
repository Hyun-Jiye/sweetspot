```python
import pandas as pd 

population = pd.read_csv('./서울 지하철역 생활인구.csv' , encoding = 'utf-8')
```

```python
population['MZ세대'] = population['연령대_30_유동인구_수'] + population['연령대_20_유동인구_수']
```

```python
population
```

```python
population = population.sort_values(by = ['MZ세대'] , ascending= False)
```

```python
population.to_csv('./MZ세대 포함 유동인구.csv', index = False, encoding = 'utf-8-sig')
```

```python
grouped_df = population.groupby('상권_코드_명')['MZ세대'].sum().reset_index()
```

```python
top_50_station = grouped_df.sort_values(by = 'MZ세대', ascending=False)
```

```python
top_50_station # MZ 세대가 가장 많은 상권(역)
```

```python
near_yaksu = ['동대입구역', '금호역', '신당역','버티고개역']
prof_df = top_50_station['상권_코드_명'].isin(near_yaksu)

prof_df.any()
```

```python
top_100_stations = grouped_df.sort_values(by='MZ세대', ascending=False)
top_100_stations = top_100_stations.head(100)
prof_df_100 = top_100_stations['상권_코드_명'].isin(near_yaksu)
prof_df_100.any()
```

```python
top_100_stations.to_csv('./top_100_stations.csv',index=False, encoding='utf-8-sig')
```
