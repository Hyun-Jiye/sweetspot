```python
import pandas as pd
```

### 서울 지하철역 데이터 불러오기

```python
subway = pd.read_csv('data/상권 관련/서울시 지하철역 정보 검색 (역명).csv', encoding='cp949', index_col=None)
```

```python
subway
```

### 전철역명 데이터 끝자리에 '역' 추가

```python
subway['전철역명'] = subway['전철역명'].apply(lambda x: x + '역' if not x.endswith('역') else x)
```

```python
subway
```

### 서울 생활인구 데이터 불러오기

```python
population = pd.read_csv('data/상권 관련/서울시 상권분석서비스(길단위인구-상권).csv', encoding='cp949', index_col=None)
```

```python
population
```

### 생활인구 데이터에서 지하철역에 해당하는 데이터만 필터링

```python
# subway 데이터프레임의 '전철역명'을 리스트로 변환
subway_station_list = subway['전철역명'].tolist()

# '상권_코드_명' 컬럼에서 subway 데이터프레임의 '전철역명'을 포함하는 행을 필터링
seoul = population[population['상권_코드_명'].apply(lambda x: any(station in x for station in subway_station_list))]
```

```python
seoul
```

### 지하철역 및 출구에 해당하지 않는 데이터 확인 (전통시장)

```python
seoul[seoul['상권_구분_코드_명'] == '전통시장']
```

### 전통시장 제외

```python
seoul = seoul[seoul['상권_구분_코드_명'] != '전통시장']
```

```python
seoul
```
