```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns 
```

```python
df = pd.read_csv('./젠트리피케이션_소팅.csv', encoding= 'utf-8')
df.head()
```

```python
gr_df = df.groupby('상권_코드_명').apply(lambda x : x.sort_values('기준_년분기_코드')).reset_index(drop=True)
print(gr_df)
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt 
import matplotlib as mpl

mpl.rcParams['font.family'] = 'AppleGothic'
mpl.rcParams['axes.unicode_minus'] = False

plt.figure(figsize = (10,8))
sns.scatterplot(data = gr_df, x = '기준_년분기_코드', y = '상권_변화_지표',hue = '상권_코드_명', style = '상권_코드_명', s = 100)
plt.xticks(rotation = 45)
plt.title('젠트리피케이션 신당')
plt.show()
```
