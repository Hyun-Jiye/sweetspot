```python
def geocode(address):
    base_url = "https://maps.googleapis.com/maps/api/geocode/json"
    params = {
            'address' : address + ', Seoul, South Korea',
            'key' : API_KEY
    }

    response = requests.get(base_url, params=params)
    print(response)
    if response.status_code == 200:
        json_response = response.json()
        if json_response['results']:
            location = json_response['results'][0]['geometry']['location']
            return (location['lat'], location['lng'])
        
    return (None, None)


#Dataframe 위도 경도 추가
df['위도'], df['경도'] = zip(*df['행정동'].map(geocode))
df = df.dropna(subset = ['위도','경도']) # 유효하지 않은 위치 데이터는 제거 

#center lat, lng설정 
center_lat = df['위도'].mean()
center_lng = df['경도'].mean()

#Folium생성
map = folium.Map(location=[center_lat,center_lng], zoom_start = 12)


# heatmap으로 표현하여 시각적으로 보기 위해 표현. 
heat_data = [[row['위도'], row['경도'], row['MZ세대_유동인구']] for idx , row in df.iterrows()]
heat_map = plugins.HeatMap(heat_data, radius=40, max_zoom=13, blur=15,max_val = 1.0)
map.add_child(heat_map)


# 잘 보이지 않으니 마커를 추가하자
for idx , row in df.iterrows():
    if row['위도'] is not None and row['경도'] is not None:
        folium.Marker(
            location = (row['위도'], row['경도']),
            popup = f'{row["행정동"]} : 유동인구 {row["MZ세대_유동인구"]}',
            icon = folium.Icon(color = 'blue', icon = 'info-sign')
                    
        ).add_to(map)
    else :
        print(row['행정동'])


#지도저장 


map.save('heatmap.html')
```
