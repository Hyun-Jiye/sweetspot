```python
df = pd.read_csv('./MZ세대 포함 유동인구.csv', encoding='utf-8')
```

```python
df[df['상권_코드_명'] == '동대입구역'] 
```

### 약수역 포함 약수역 인근 지하철역에 대한 list 

```python
near_yaksu_staitons = ['약수역','동대입구역','신당역','금호역','버티고개역']

```

```python
filtered_df = df[df['상권_코드_명'].apply(lambda x: any(station in x for station in near_yaksu_staitons))]

filtered_df
```

```python
anal_pop_individual = filtered_df.groupby(['상권_코드_명', '기준_년분기_코드'])['MZ세대'].sum().unstack()
```

```python
anal_pop_individual
```

```python
# 약수역 인근 역의 유동인구를 전체 햡쳐서 분기별 기준으로 보는 경우 

anal_pop_tot = filtered_df.groupby('기준_년분기_코드')['MZ세대'].sum()

print(anal_pop_tot)
```

```python
import matplotlib.pyplot as plt 
```

```python
plt.figure(figsize = (10,6))

anal_pop_tot.plot(kind = 'bar', color = 'skyblue')
plt.title('Quarterly Population Trend')
plt.xlabel('Q')
plt.ylabel('Total Population near yaksu')
plt.xticks(rotation =45)
plt.grid(True, linestyle = '--', alpha = 0.6)
plt.show()
```

#### 직관적으로 보고자 plotly 를 사용하였으나 보기가 더 어렵다. 

```python
# 선형 추세선 적용하여 조금 더 직관적으로 보기 위해 
from sklearn.linear_model import LinearRegression
```

```python
fig, axes = plt.subplots(nrows= len(anal_pop_individual.index), ncols =1, figsize = (10,5 * len(anal_pop_individual.index)) , sharex = True  )


for i , (index , row) in enumerate(anal_pop_individual.iterrows()):
    X = np.array(range(len(row.index))).reshape(-1,1)

    # 선형 회귀 모델 적용하는 코드
    model = LinearRegression()
    model.fit(X, row.values)

    trend_line = model.predict(X)


    axes[i].plot(row.index, row, marker = 'o', linestyle= '-', label = f'{index} MZ세대 유동인구')
    axes[i].plot(row.index, trend_line,  color = 'red', linestyle = '--', label = '추세선')

    axes[i].set_title(f'{index} MZ세대 유동인구 추세')
    axes[i].set_ylabel('MZ세대 유동인구 수')
    axes[i].grid(True)
    axes[i].legend()


axes[-1].set_xlabel('기준_년분기_코드')

plt.tight_layout()
plt.show()


```

#### 코드 설명 

- plt.subplots 를 통해 각각의 추이를 자세히 들여다보고자 하였음.
#### nrows = len(anal_pop_individual.index) , ncols = 1, figsize = (10, 5 * len(anal_pop_individual.index))
- 서브플롯의 행 수를 데이터 프레임의 행 수 (각각의 상권 수) 만큼 설정하고 , 상권별로 하나의 서브플롯 생성
- 열 수를 1로 설정해서 모든 서브플롯은 하나의 열로 배열된다. 
- sharex를 통해서 모든 서브플롯이 x축을 공유하도록 설정한다. 



#### 증감의 추이를 조금 더 명확하게 보기 위해 LinearRegression model을 활용하여 추세선 추가 
- 분기별 데이터를 통해 데이터 포인트들과 선 사이의 거리(오차)를 최소화하는 방식으로 작동하여 (최소제곱법) 모든 데이터 포인트들에 대한 선의 수직 거리(잔차) 의 제곱합을 최소화한다. 
- 모델을 통해, 추세선을 그려 증가인지 감소인지 직관적으로 바라볼 수 있음.
