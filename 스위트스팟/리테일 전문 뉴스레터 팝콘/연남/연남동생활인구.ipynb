## 데이터 핸들링

### 일자+시간대별로 존재하는 데이터를 일자별로 나타내기 (일자를 기준으로 합계)

```python
total = round(data.groupby('기준일ID')['총생활인구수'].sum())
total
```

```python
yeonnam = pd.DataFrame(total)
yeonnam
```

```python
yeonnam = yeonnam.reset_index()
yeonnam
```

### 일자를 datetime 형식으로 변경

```python
yeonnam['기준일ID'] = pd.to_datetime(yeonnam['기준일ID'], format='%Y%m%d')
yeonnam
```

### 연도, 분기 컬럼 생성 (dt 함수 사용)

```python
yeonnam['연도'] = yeonnam['기준일ID'].dt.year
yeonnam['분기'] = yeonnam['기준일ID'].dt.quarter
yeonnam
```

### 연도+분기 기준 평균 생활인구수 집계

```python
result = round(yeonnam.groupby(['연도','분기'])['총생활인구수'].mean())
result
```

```python
result = pd.DataFrame(result).reset_index()
result
```

```python
result.to_csv("서교동_연남동_생활인구.csv", index=False, encoding='utf-8-sig')
```
