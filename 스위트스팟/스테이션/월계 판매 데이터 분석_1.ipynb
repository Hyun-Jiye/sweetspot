```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm
from matplotlib.ticker import ScalarFormatter
```

### 데이터 불러오기

```python
df = pd.read_excel('data/스테이션/스테이션 월계점 판매 데이터 (24.01~04)/스테이션_월계점.xlsx')
```

```python
df
```

### 브랜드별 판매금액

```python
brand_sales_sum = df.groupby('브랜드명')['판매금액'].sum()
brand_sales_sum = pd.DataFrame(brand_sales_sum)
brand_sales_sum.reset_index(inplace=True)
brand_sales_sum
```

```python
# 한글 폰트 설정
plt.rc('font', family='Malgun Gothic')
```

```python
# 막대그래프로 시각화
brand_sales_sum.plot(kind='bar', color='skyblue')
plt.xlabel('브랜드명')  # x축 라벨
plt.ylabel('판매금액')  # y축 라벨
plt.title('브랜드별 판매금액')  # 그래프 제목
plt.xticks(rotation=0)  # x축 라벨의 회전 각도
plt.show()
```

```python
brand_sales_count = df.groupby('브랜드명')['판매수량'].sum()
brand_sales_count = pd.DataFrame(brand_sales_count)
brand_sales_count.reset_index(inplace=True)
brand_sales_count
```

### 데이터 타입 변경 (판매일자)

```python
df['판매일자'] = pd.to_datetime(df['판매일자'])
```

```python
df.dtypes
```

### 판매금액 추이 시각화

```python
# 일별 매출 금액 합계 계산
sales_sum = df.groupby('판매일자')['판매금액'].sum()

# 시계열 꺾은선 그래프로 시각화
plt.figure(figsize=(10, 6))  # 그래프 크기 설정
plt.plot(sales_sum.index, sales_sum.values, marker='o', linestyle='-', color='blue')  # 꺾은선 그래프 생성
plt.xlabel('판매일자')  # x축 레이블
plt.ylabel('판매금액')  # y축 레이블
plt.title('스테이션 월계점 판매금액 추이(2024)')  # 그래프 제목

# Y축을 정수 형식으로 설정
y_formatter = ScalarFormatter(useOffset=False)
y_formatter.set_scientific(False)
plt.gca().yaxis.set_major_formatter(y_formatter)

plt.grid(True)  # 격자 표시
plt.xticks(rotation=45)  # x축 레이블 회전
plt.tight_layout()  # 레이아웃 조정
plt.show()  # 그래프 표시
```

```python
# 막대그래프로 시각화
plt.figure(figsize=(10, 6))  # 그래프 크기 설정
plt.bar(sales_sum.index, sales_sum.values, color='skyblue')  # 막대그래프 생성
plt.xlabel('판매일자')  # x축 레이블
plt.ylabel('판매금액')  # y축 레이블
plt.title('스테이션 월계점 판매금액 추이(2024)')  # 그래프 제목

# Y축을 정수 형식으로 설정
y_formatter = ScalarFormatter(useOffset=False)
y_formatter.set_scientific(False)
plt.gca().yaxis.set_major_formatter(y_formatter)

plt.grid(True)  # 격자 표시
plt.xticks(rotation=45)  # x축 레이블 회전
plt.tight_layout()  # 레이아웃 조정
plt.show()  # 그래프 표시
```
