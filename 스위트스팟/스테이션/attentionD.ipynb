```python
import pandas as pd
```

```python
data1 = pd.read_csv('data/인구 관련/유동인구 측정 기기/어텐션디.csv', encoding='UTF8')
data2 = pd.read_csv('data/인구 관련/유동인구 측정 기기/딥라운지.csv', encoding='UTF8')
```

```python
data1
```

```python
data2
```

```python
a = data1.groupby(['날짜','요일','시간','성별','연령'])['방문수'].sum()
a = pd.DataFrame(a)
a.reset_index(inplace=True)
a
```

```python
a.rename(columns={'방문수':'방문수_어텐션디'}, inplace=True)
a
```

```python
a1 = data1.groupby(['날짜','요일','시간'])['reach'].sum()
a1 = pd.DataFrame(a1)
a1.reset_index(inplace=True)
a1
```

```python
a1.rename(columns={'reach':'방문수_어텐션디'}, inplace=True)
a1
```

```python
b = data2.groupby(['날짜','요일','시간','성별','연령'])['방문수'].sum()
b = pd.DataFrame(b)
b.reset_index(inplace=True)
b
```

```python
b.rename(columns={'방문수':'방문수_딥라운지'}, inplace=True)
b
```

```python
b1 = data2.groupby(['날짜','요일','시간'])['방문수'].sum()
b1 = pd.DataFrame(b1)
b1.reset_index(inplace=True)
b1
```

```python
b1.rename(columns={'방문수':'방문수_딥라운지'}, inplace=True)
b1
```

```python
total = pd.merge(a, b, on=['날짜','요일','시간','성별','연령'], how='outer')
total
```

```python
total1 = pd.merge(a1, b1, on=['날짜','요일','시간'], how='outer')
total1 = total1.sort_values(by=['날짜','시간'])
total1
```

```python
total1.to_csv("어텐션디_딥라운지.csv", index=False, encoding='utf-8-sig')
```

```python
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm

# 한글 폰트 설정
plt.rc('font', family='Malgun Gothic') 

# '날짜' 컬럼을 날짜 형식으로 변환
total1['날짜'] = pd.to_datetime(total1['날짜'])

# 날짜를 기준으로 데이터프레임 정렬
total1 = total1.sort_values(by='날짜')

# 꺾은선 그래프 그리기
plt.figure(figsize=(10, 6)) # 그래프 크기 설정
plt.plot(total1['날짜'], total1['방문수_어텐션디'], label='방문수_어텐션디', linestyle='-') # 첫 번째 꺾은선 그래프
plt.plot(total1['날짜'], total1['방문수_딥라운지'], label='방문수_딥라운지', linestyle='--') # 두 번째 꺾은선 그래프

plt.title('날짜별 방문수 비교') # 그래프 제목
plt.xlabel('날짜') # x축 라벨
plt.ylabel('방문수') # y축 라벨
plt.legend() # 범례 표시
plt.xticks(rotation=45) # x축 라벨 회전
plt.grid(True) # 그리드 표시

plt.show() # 그래프 표시
```

```python
from scipy.stats import ttest_ind
import numpy as np

# 방문수_어텐션디와 방문수_딥라운지 컬럼의 결측값을 0으로 채움
total1['방문수_어텐션디'].fillna(0, inplace=True)
total1['방문수_딥라운지'].fillna(0, inplace=True)

# t-test 수행
t_stat, p_value = ttest_ind(total1['방문수_어텐션디'], total1['방문수_딥라운지'])

print(f'T-statistic: {t_stat}')
print(f'P-value: {p_value}')
```

```python
test = total1[total1['날짜']>='2024-02-24']
test
```

```python
# t-test 수행
t_stat, p_value = ttest_ind(test['방문수_어텐션디'], test['방문수_딥라운지'])

print(f'T-statistic: {t_stat}')
print(f'P-value: {p_value}')
```

```python
from scipy.stats import mannwhitneyu

# Mann-Whitney U 검정을 수행합니다.
u_stat, p_value = mannwhitneyu(test['방문수_어텐션디'], test['방문수_딥라운지'], alternative='two-sided')

print("U-statistic:", u_stat)
print("p-value:", p_value)

# p-value의 해석:
# p-value가 0.05(또는 선택한 유의 수준)보다 작으면, 두 집단 간에는 통계적으로 유의미한 차이가 있다고 할 수 있습니다.
# p-value가 0.05보다 크면, 두 집단 간에 통계적으로 유의미한 차이가 없다고 할 수 있습니다.
# 이 경우, 두 회사의 유동인구 측정 데이터 간에 유의미한 차이가 없다면 정합성이 있다고 볼 수 있습니다.
```
