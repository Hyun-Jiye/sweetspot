```python
import pandas as pd 

data =pd.read_csv('광고EDA.csv', encoding='utf-8-sig')

data.info()
```

```python
data
```

```python
columns_to_clean = ['지출금액', 'CTR(전체)','CPC(전체)','CPM','결과 당 비용','고객 획득 당 비용']

for col in columns_to_clean :
    if data[col].dtype == 'object':
        data[col] = data[col].replace('[\%,₩]','',regex=True).astype(float)

exclude_ads = ['14안', '15안','16안','17안','18안']
data = data[~data['소재'].isin(exclude_ads)]


```

```python
data['지출대비전환'] = data['고객 전환'] / data['지출금액']


good_performance_threshold = data['지출대비전환'].mean()

data['cluster'] = 'Medium'

data.loc[data['지출대비전환'] >= good_performance_threshold, 'cluster'] = 'Good'
data.loc[data['지출대비전환'] < good_performance_threshold, 'cluster'] = 'Bad'

```

```python
good_summary = data[data['cluster'] == 'Good'].describe()
bad_summary = data[data['cluster'] == 'Bad'].describe()

print("Good Cluster Summary:\n", good_summary)
print("Bad Cluster Summary :\n", bad_summary)

```

```python
pop_con_channel = data[data['매체 '] == '팝콘']
pop_going_channel = data[data['매체 '] == '팝업스토어가자']

```

```python
pop_con_stats = pop_con_channel[['지출금액', 'CTR(전체)', 'CPC(전체)', 'CPM', '고객 전환']].describe()
pop_going_stats = pop_going_channel[['지출금액', 'CTR(전체)', 'CPC(전체)', 'CPM', '고객 전환']].describe()

print("팝콘 채널 기술 통계량 : \n", pop_con_stats)
print("팝업스토어 가자 채널 기술 통계량 : \n", pop_going_stats)

```

```python
# 채널 군집 나누어서 상관 행렬 

def standardize_and_corr(data):
    metrics = data[['지출금액', 'CTR(전체)','CPC(전체)','CPM', '고객 전환']]
    scaler = StandardScaler()
    metrics_scaled = scaler.fit_transform(metrics)
    metrics_scaled = pd.DataFrame(metrics_scaled, columns = metrics.columns)
    return metrics_scaled.corr()
```

```python
pop_con_corr_metrics = standardize_and_corr(pop_con_channel)
pop_going_corr_metrics = standardize_and_corr(pop_going_channel)
```

```python
plt.figure(figsize = (10,8))

plt.subplot(1,2,1)
sns.heatmap(pop_going_corr_metrics , annot = True, cmap = 'coolwarm')
plt.title('팝콘 채널 상관 행렬')

plt.subplot(1,2,2)
sns.heatmap(pop_going_corr_metrics, annot = True, cmap ='coolwarm')
plt.title('팝업스토어가자 채널 상관 행렬')

plt.tight_layout()
plt.show()
          
```

```python
## 채널의 영향력 

popcon_conversion_rate = pop_con_channel['고객 전환'].sum() / pop_con_channel['지출금액'].sum()
pop_going_conversion_rate = pop_going_channel['고객 전환'].sum() / pop_going_channel['지출금액'].sum()

print('팝콘 채널 고객 전환율: ' , popcon_conversion_rate)
print('팝업스토어 가자 채널 고객 전환율: ',pop_going_conversion_rate)
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(x=['팝콘', '팝업스토어 가자'], y=[popcon_conversion_rate, pop_going_conversion_rate],color = 'green')
plt.title('고객 전환율 비교')
plt.ylabel('고객 전환율')
plt.show()
```

```python
from scipy.stats import ttest_ind

t_stat, p_value = ttest_ind(popcon_conversion_rate, pop_going_conversion_rate, nan_policy='omit')

print("t-statistic: ", t_stat)
print("p-value :" , p_value)
```

```python
print(pop_con_channel[['고객 전환', '지출금액']].isna().sum())
print(pop_going_channel[['고객 전환', '지출금액']].isna().sum())

print(pop_con_channel[['고객 전환', '지출금액']].eq(0).sum())
print(pop_going_channel[['고객 전환', '지출금액']].eq(0).sum()) # 0 값은 존재
```

```python
# 0 값 제거 후 다시 한번 전환율 체크 및 통계량 분석 
```

```python
pop_con_channel = pop_con_channel.dropna(subset = ['고객 전환', '지출금액'])
pop_going_channel = pop_going_channel.dropna(subset = ['고객 전환', '지출금액'])
```

```python
pop_con_channel = pop_con_channel[(pop_con_channel['고객 전환'] != 0) & (pop_con_channel['지출금액'] != 0)]
pop_going_channel = pop_going_channel[(pop_going_channel['고객 전환']!= 0) & (pop_going_channel['지출금액']!= 0)]

```

```python
# 전환율 
pop_con_channel['전환율'] = pop_con_channel['고객 전환'] / pop_con_channel['지출금액']
pop_going_channel['전환율'] = pop_going_channel['고객 전환'] / pop_going_channel['지출금액']

popcon_conversion_rate = pop_con_channel['전환율']
popgoing_conversion_rate = pop_going_channel['전환율']

print(popcon_conversion_rate.mean())
print(popgoing_conversion_rate.mean())
```

```python
from scipy.stats import ttest_ind
# 독립표본 t-test
# t-검정 수행
t_stat, p_value = ttest_ind(popcon_conversion_rate, popgoing_conversion_rate, nan_policy='omit')

print("t-statistic: ", t_stat)
print("p-value: ", p_value)

```

```python
from scipy.stats import ttest_rel

pop_con_conversion_rate = pop_con_channel.set_index('소재')['전환율']
pop_going_conversion_rate = pop_going_channel.set_index('소재')['전환율']

common_ads = pop_con_conversion_rate.index.intersection(pop_going_conversion_rate.index)
pop_con_conversion_rate = pop_con_conversion_rate.loc[common_ads]
pop_going_conversion_rate = pop_going_conversion_rate.loc[common_ads]

# 대응 표본 t-test 수행
t_stat, p_value = ttest_rel(pop_con_conversion_rate, pop_going_conversion_rate)

print("t-statistic: ", t_stat)
print("p-value: ", p_value)
```

- 광고 성과 좋은 집단(Good Performance)

1. 지출금액과 고객 전환 (0.99) : 지출금액과 고객 전환 간의 매우 강한 양의 상관관계를 보인다. 이는 광고에 더 많은 돈을 지출할수록 고객 전환이 크게 증가함을 의미한다. 
2. CTR (전체) 와 고객 전환 (0.30) : 클릭률이 높을 수록 고객전환이 증가하는 경향성이 있으나 이러한 상관관계의 정도는 중간 정도이다. 
3. CPC 와의 고객전환 : 클릭 당 비용이 낮을 수록 고객 전환이 증가하는 경향성을 보인다. 
4. CPM 과 고객전환 : 약한 음의 상관관계를 보이는 것으로 보아 CTR 와 CPC 보다는 지표의 정도가 강하지 않다. 천회 노출당 비용이 낮을 수록 고객 전환이 증가하는 경향이 있다. 

- 광고 성과 안 좋은 집단 (Bad Performance) 
1. 지출금액과 고객 전환 (0.53) : 지출이 증가할 수록 고객 전환이 증가하나, 상관관계의 경우 성과가 좋은 집단에 비해 약한 상관관계를 보인다. 
2. CTR (전체) 와 고객 전환 (0.11)
3. CPC , 고객전환 (-0.02) 


Insight 
지표 관련하여 
- 예산 할당이 매우 중요한 요소 
- 클릭률을 높인다면 전환율 개선에 기여 할 수 있다. 이와 동일하게 CPC 클릭 당 비용을 줄여 전환율을 높이는 것에 집중해야 한다. 


채널 간의 영향력 관련하여 
- 가설 : 두 채널 간의 영향력 차이가 없다.(리드 확보에 있어)
- 가시적으로 확인 한 결과 상대적으로 팝콘 계정의 리드 확보가 더 높지만 현재 사용한 통계 분석으로는 유의미한 차이를 보이지 않았습니다. 다양한 지표를 고려하여 유의미한 차이가 있는지 검정해볼 생각입니다. 
    이를 통해, 이후 행보에서 데이터를 더욱 확보하고 채널 간 A/B test (동일 소재)를 통해 우리의 예산을 어디로 더 많이 할당하여 전환율을 높일 수 있을지에 대한 실험을 많이 진행해야 할 것 같습니다.
