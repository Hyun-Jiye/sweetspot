```python
import pandas as pd
```

```python
data1 = pd.read_csv('data/매장 관련/서울시 일반음식점 인허가 정보.csv', encoding='utf-8-sig')
data2 = pd.read_csv('data/매장 관련/서울시 휴게음식점 인허가 정보.csv', encoding='utf-8-sig')
data3 = pd.read_csv('data/매장 관련/서울시 제과점영업 인허가 정보.csv', encoding='utf-8-sig')
```

```python
d1 = pd.concat([data1, data2], axis=0)
d2 = pd.concat([d1, data3], axis=0)
```

```python
d2
```

```python
d2022 = d2[d2['인허가일자'].between('2022-01-01', '2022-12-31')]
d2022.sort_values('인허가일자', ascending=True, inplace=True)
d2022.reset_index()
```

```python
d2023 = d2[d2['인허가일자'].between('2023-01-01', '2023-12-31')]
d2023.sort_values('인허가일자', ascending=True, inplace=True)
d2023.reset_index()
```

```python
d2022.to_csv("성수동_신규매장_2022년.csv", index=False, encoding='utf-8-sig')
d2023.to_csv("성수동_신규매장_2023년.csv", index=False, encoding='utf-8-sig')
```
