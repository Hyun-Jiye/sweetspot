```python
import pandas as pd
```

### 모든 지역 데이터 불러오기

```python
data1 = pd.read_csv('data/트위터/twitter_강남구.csv', encoding='utf-8-sig')
data2 = pd.read_csv('data/트위터/twitter_강동구.csv', encoding='utf-8-sig')
data3 = pd.read_csv('data/트위터/twitter_강북구.csv', encoding='utf-8-sig')
data4 = pd.read_csv('data/트위터/twitter_강서구.csv', encoding='utf-8-sig')
data5 = pd.read_csv('data/트위터/twitter_관악구.csv', encoding='utf-8-sig')
data6 = pd.read_csv('data/트위터/twitter_광진구.csv', encoding='utf-8-sig')
data7 = pd.read_csv('data/트위터/twitter_구로구.csv', encoding='utf-8-sig')
data8 = pd.read_csv('data/트위터/twitter_동대문구.csv', encoding='utf-8-sig')
data9 = pd.read_csv('data/트위터/twitter_동작구.csv', encoding='utf-8-sig')
data10 = pd.read_csv('data/트위터/twitter_마포구.csv', encoding='utf-8-sig')
data11 = pd.read_csv('data/트위터/twitter_서대문구.csv', encoding='utf-8-sig')
data12 = pd.read_csv('data/트위터/twitter_서초구.csv', encoding='utf-8-sig')
data13 = pd.read_csv('data/트위터/twitter_성동구.csv', encoding='utf-8-sig')
data14 = pd.read_csv('data/트위터/twitter_성북구.csv', encoding='utf-8-sig')
data15 = pd.read_csv('data/트위터/twitter_송파구.csv', encoding='utf-8-sig')
data16 = pd.read_csv('data/트위터/twitter_양천구.csv', encoding='utf-8-sig')
data17 = pd.read_csv('data/트위터/twitter_영등포구.csv', encoding='utf-8-sig')
data18 = pd.read_csv('data/트위터/twitter_용산구.csv', encoding='utf-8-sig')
data19 = pd.read_csv('data/트위터/twitter_은평구.csv', encoding='utf-8-sig')
data20 = pd.read_csv('data/트위터/twitter_중구.csv', encoding='utf-8-sig')
data21 = pd.read_csv('data/트위터/twitter_중랑구.csv', encoding='utf-8-sig')
```

### 데이터프레임 합치기

```python
d1 = pd.concat([data1, data2], axis=0)
d2 = pd.concat([d1, data3], axis=0)
d3 = pd.concat([d2, data4], axis=0)
d4 = pd.concat([d3, data5], axis=0)
d5 = pd.concat([d4, data6], axis=0)
d6 = pd.concat([d5, data7], axis=0)
d7 = pd.concat([d6, data8], axis=0)
d8 = pd.concat([d7, data9], axis=0)
d9 = pd.concat([d8, data10], axis=0)
d10 = pd.concat([d9, data11], axis=0)
d11 = pd.concat([d10, data12], axis=0)
d12 = pd.concat([d11, data13], axis=0)
d13 = pd.concat([d12, data14], axis=0)
d14 = pd.concat([d13, data15], axis=0)
d15 = pd.concat([d14, data16], axis=0)
d16 = pd.concat([d15, data17], axis=0)
d17 = pd.concat([d16, data18], axis=0)
d18 = pd.concat([d17, data19], axis=0)
d19 = pd.concat([d18, data20], axis=0)
d20 = pd.concat([d19, data21], axis=0)
```

```python
d20
```

### 중복 트윗 제거

```python
data = d20.drop_duplicates()
```

```python
data
```

### 데이터프레임 csv로 저장

```python
data.to_csv("트위터_통합.csv", index=False, encoding='utf-8-sig')
```
